Get cluster size distributions for every peak chosen per month

```python
after_nonflash = pd.read_json('output_clusters/after_clusters_nonflash.jsonl', lines=True)
after_flash = pd.read_json('output_clusters/after_clusters_flash.jsonl', lines=True)
after_flash['num_clusters'] = after_flash['clusters'].apply(lambda x: len(x))
after_nonflash['num_clusters'] = after_nonflash['clusters'].apply(lambda x: len(x))
# concatenate dfs
df_after = pd.concat([after_nonflash, after_flash])   
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_after['num_clusters'], bins=range(1, df_after['num_clusters'].max() + 2), edgecolor='black', align='left')

# Add labels and title
plt.xlabel('Number of Clusters')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Clusters')

# Show the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(range(1, df_after['num_clusters'].max() + 1))
plt.show()
```

```python
during_nonflash = pd.read_json('output_clusters/during_clusters_nonflash.jsonl', lines=True)
during_nonflash['num_clusters'] = during_nonflash['clusters'].apply(lambda x: len(x))
during_flash = pd.read_json('output_clusters/during_clusters_flash.jsonl', lines=True)
during_flash['num_clusters'] = during_flash['clusters'].apply(lambda x: len(x))
df_during = pd.concat([during_nonflash, during_flash])   
```

```python
import matplotlib.pyplot as plt

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Plot for df_before
axes[0].hist(df_before['num_clusters'], bins=range(1, df_before['num_clusters'].max() + 2), 
             edgecolor='black', align='left')
axes[0].set_xlabel('Number of Clusters')
axes[0].set_title('Before Distribution')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)
axes[0].set_xticks(range(1, df_before['num_clusters'].max() + 1))

# Plot for df_during
axes[1].hist(df_during['num_clusters'], bins=range(1, df_during['num_clusters'].max() + 2), 
             edgecolor='black', align='left')
axes[1].set_xlabel('Number of Clusters')
axes[1].set_title('During Distribution')
axes[1].grid(axis='y', linestyle='--', alpha=0.7)
axes[1].set_xticks(range(1, df_during['num_clusters'].max() + 1))

# Plot for df_after
axes[2].hist(df_after['num_clusters'], bins=range(1, df_after['num_clusters'].max() + 2), 
             edgecolor='black', align='left')
axes[2].set_xlabel('Number of Clusters')
axes[2].set_ylabel('Frequency')
axes[2].set_title('After Distribution')
axes[2].grid(axis='y', linestyle='--', alpha=0.7)
axes[2].set_xticks(range(1, df_after['num_clusters'].max() + 1))

# Adjust layout for better spacing
plt.tight_layout()

# Show the combined plot
# save image
plt.savefig('output_clusters/distributions_numclusters.png')
plt.show()
```
